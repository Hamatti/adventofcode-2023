## Read input

Today's input is a single line with comma separated values. I strip out the possible newline in the input and split by comma. Final sequence is then the first value in all the lines.

````python
from utils import read_input

def transformer(line):
    return line.strip().split(',')

sequence = read_input(15, transformer)[0]
````

Calculation of the hash is step-by-step implementation from the instructions. Python's built-in [ord()](https://docs.python.org/3/library/functions.html#ord) function gives us the ASCII value for a given character.

````python
def calculate_hash(word):
    hash_value = 0
    for ch in word:
        hash_value = hash_value + ord(ch)
        hash_value = hash_value * 17
        hash_value = hash_value % 256
    return hash_value
````

To calculate the final result, we calculate hash for each step and sum them up.

````python
part_1 = 0
for step in sequence:
    part_1 += calculate_hash(step)
        
print(f'Solution: {part_1}')
assert part_1 == 517965
````

The hardest part today for me was reading. Despite multiple reads, I missed that in the second part we're supposed to hash only the label part (the part before `=`/`-`).

In my solution, I take advantage of the fact that since Python 3.7, dictionaries are guaranteed to return items in insertion order. I initially had a `OrderedDict` since I wasn't quite sure if this was guarantee as some documentation has said it's like this for now but that it's an implementation detail that can change.

But reading through [OrderedDict](https://docs.python.org/3/library/collections.html#ordereddict-objects) documentation, it says:

> Ordered dictionaries are just like regular dictionaries but have some extra capabilities relating to ordering operations. They have become less important now that the built-in dict class gained the ability to remember insertion order (this new behavior became guaranteed in Python 3.7).

and I'll trust that for now.

My HASHMAP is a dictionary with dictionaries inside. In the `lenses` dictionary, I store the label and its corresponding focal length. This makes replacing and deleting lenses efficient compared to for example storing them in a list or a tuple.

To create the HASHMAP, I go through each step in sequence and depending on the operator (`=` or `-`), add or remove it from the HASHMAP. To calculate the focusing power, I iterate over each box and all the lenses in that box and calculate power for each lense based on the formula provided in the puzzle and sum them all up.

````python
def add_to_hashmap(step, hashmap):
    label, focal_length = step.split('=')
    
    key = calculate_hash(label)
    lenses = hashmap.get(key, {})
    
    lenses[label] = int(focal_length)
    hashmap[key] = lenses
    
def delete_from_hashmap(step, hashmap):
    label = step.split('-')[0]

    key = calculate_hash(label)
    lenses = hashmap.get(key, {})
    
    if label in lenses:
        del lenses[label]
    hashmap[key] = lenses

def create_hashmap(sequence):
    hashmap = {}
    for step in sequence:
        if '=' in step:
            add_to_hashmap(step, hashmap)
        if '-' in step:
            delete_from_hashmap(step, hashmap)
    return hashmap

def calculate_focusing_power(hashmap):
    focusing_power = 0
    for box, lenses in hashmap.items():
        for slot, (label, focal_length) in enumerate(lenses.items(), start=1):
            focusing_power += (box + 1) * slot * focal_length
    return focusing_power
````

````python
hashmap = create_hashmap(sequence)
part_2 = calculate_focusing_power(hashmap)

print(f'Solution: {part_2}')
assert part_2 == 267372
````
